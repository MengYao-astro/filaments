```python
import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits
```

```python
def coldsty(vr,vNH2c,vRflat,vp):                      # NH2
    return vNH2c/np.power(1.+(vr/vRflat)**2.,(vp-1.)/2.)
NH2c=10.**23
r=np.linspace(-50,50,101)


NFLMTs = np.zeros((3,101,101))
for diff_NH2 in 1.,2.,3. :
    for i in range (101):
        Ap=np.pi/2.
        Rflat=10.
        j=int(diff_NH2-1.)     #count from 0
        p=2.
        NFLMTs[j,i,:]=coldsty(r,diff_NH2*10**23,Rflat,p)     #write column density to 3 arrays
    plt.subplot(1,3,1)    
    plt.plot(r,NFLMTs[j,0,:])     #plot with r and each column desity
    plt.title('diff p')
    plt.legend(['2','3','4'],loc='best')        

RflatFLMTs=np.zeros((3,101,101))
for Rflat in 10.,20.,30.:
    for i in range(101):
        Ap=np.pi/2.
        p=2.
        j=int(Rflat/10.-1)    #count from 0
        RflatFLMTs[j,i,:]=coldsty(r,NH2c,Rflat,p)   #write column density to 3 arrays
    plt.subplot(1,3,2)
    plt.plot(r,RflatFLMTs[j,0,:])    #plot with r and each column desity
    plt.title('diff Rflat')
    plt.legend(['10','20','30'],loc='best')
#diff p
pFLMTs=np.zeros((3,101,101))
for p in np.arange(2,5,1):
    for i in range(101):
        Ap=np.pi/2.
        Rflat=10.
        j=int(p-2.)     #count from 0
        pFLMTs[j,i,:]=coldsty(r,NH2c,Rflat,p)     #write column density to 3 arrays
    plt.subplot(1,3,3)    
    plt.plot(r,pFLMTs[j,0,:])     #plot with r and each column desity
    plt.title('diff p')
    plt.legend(['2','3','4'],loc='best')
plt.tight_layout()
plt.show()
```
